# Kosac 사전 불러오기

```python
kosac = pd.read_csv('Kosac/polarity.csv')
kosac
```

```python
kosac = kosac.drop(columns=['freq','COMP','None','max.value','max.prop'])
kosac
```

# Kosac ngram처리

```python
import re

# 텍스트 정제 함수 
def text_cleaning(text):
    # 한글 정규식 정의(띄어쓰기, ㄱ~ㅣ, 가~힣)
    hangul = re.compile('[^ ㄱ-ㅣ가-힣0-9?!.,]+')
    # 정제한다(정규식에 해당하지 않는 문자는 길이가 0인 문자열로 변환한다)
    result = hangul.sub('', text)
    return result
```

```python
for i in kosac['ngram']:
    print(text_cleaning(i))
```

# KNU negative

```python
li = []
f = open("data/negative.txt", "r",encoding='UTF8')
for x in f:
    li += x.split()
```

```python
knu_neg = pd.DataFrame(li)
knu_neg.columns = ['ngram']
knu_neg['NEG'] = [1]*len(knu_neg)
knu_neg['NEUT'] = [0]*len(knu_neg)
knu_neg['POS'] = [0]*len(knu_neg)
knu_neg
```

# KNU neutral

```python
li = []
f = open("data/neutral.txt", "r",encoding='UTF8')
for x in f:
    li += x.split()
```

```python
knu_neut = pd.DataFrame(li)
knu_neut.columns = ['ngram']
knu_neut['NEG'] = [0]*len(knu_neut)
knu_neut['NEUT'] = [1]*len(knu_neut)
knu_neut['POS'] = [0]*len(knu_neut)
knu_neut
```

# KNU positive

```python
li = []
f = open("data/positive.txt", "r",encoding='UTF8')
for x in f:
    li += x.split()
```

```python
knu_pos = pd.DataFrame(li)
knu_pos.columns = ['ngram']
knu_pos['NEG'] = [0]*len(knu_pos)
knu_pos['NEUT'] = [0]*len(knu_pos)
knu_pos['POS'] = [1]*len(knu_pos)
knu_pos
```

# KNU 사전구축

```python
knu = pd.concat([knu_neg,knu_neut,knu_pos])
knu = knu.reset_index().drop(columns='index')
knu
```

```python
knu.to_csv('knu.csv',encoding='')
```
