```python
table = dict()

with open('data/dic.csv','r',-1,'utf-8') as polarity :
    next(polarity)
    
    for line in csv.reader(polarity) :
        #print(line[1].split(';'))
        key = str()
        for word in line[1].split(';'):
            key += word.split('/')[0]
            #print(key)
            
        table[key] = {'Neg':line[1], 'Neut':line[2], 'Pos':line[3]}
    print(table)
```

```python
file = pd.read_csv('data/data2030.csv')
file = file.drop('Unnamed: 0',axis=1)
file['Comment'][1]
```

```python
cnt = 20

comment_list = []
negative_list = []
netural_list = []
positive_list = []
rate_list = []

for sentence in file['Comment'][:cnt] :
    negative = 0
    netural = 0
    positive = 0
    text = splitter.morphs(sentence)
    for word in text :
        if word in table :
            print(word)
            print(table[word])
            negative += float(table[word]['Neg'])
            netural += float(table[word]['Neut'])
            positive += float(table[word]['Pos'])
            
    comment_list.append(sentence)
    negative_list.append(negative)
    netural_list.append(netural)
    positive_list.append(positive)
    if positive+negative == 0:
        rate_list.append(0)
    else:
        rate_list.append(positive/(positive+negative))
  

df['Comment'] = comment_list
df['negative'] = negative_list
df['netural'] = netural_list
df['positive'] = positive_list
df['rate'] = rate_list
```
