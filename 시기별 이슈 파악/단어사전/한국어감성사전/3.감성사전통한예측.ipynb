```python
file = pd.read_csv('data/data2030.csv')
file = file.drop('Unnamed: 0',axis=1)
file['Comment'][1]
```

# 긍정,중립,부정으로 분류한 데이터프레임 생성

```python
splitter = Okt()

cnt = 30
comment_list = []
negative_list = []
neutral_list = []
positive_list = []
rate_list = []
lex_list = enumerate(lex['ngram'])

columns=['negative','neutral','positive']
df = pd.DataFrame(columns=columns)

for sentence in file['Comment'][:cnt] :
    negative = 0
    neutral = 0
    positive = 0
    text = splitter.morphs(sentence)
    for word in text :
        # print(word)
        for idx, value in lex_list:
            print(word)
            if word in value:
                print(word)
                negative += float(lex['NEG'][idx])
                neutral += float(lex['NEUT'][idx])
                positive += float(lex['POS'][idx])
            
    comment_list.append(sentence)
    negative_list.append(negative)
    neutral_list.append(neutral)
    positive_list.append(positive)
    if positive+negative == 0:
        rate_list.append(0)
    else:
        rate_list.append(positive/(positive+negative))
  

df['Comment'] = comment_list
df['negative'] = negative_list
df['neutral'] = neutral_list
df['positive'] = positive_list
df['rate'] = rate_list
```
