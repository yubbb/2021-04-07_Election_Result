# 형용사 neg,neut,pos 분류한 데이터 프레임 생성

```python
# 형용사 긍정
li = []
for i in range(50):
    li.append(adj[i][1])
adj_pos = pd.DataFrame(li)
adj_pos.columns = ['ngram']
adj_pos['NEG'] = [0]*len(adj_pos)
adj_pos['NEUT'] = [0]*len(adj_pos)
adj_pos['POS'] = [1]*len(adj_pos)
adj_pos
```

```python
# 중립 데이터프레임 생성 위해 형용사의 중간 50개 데이터를 추출
typeTxt = adj
if len(typeTxt)/2 % 2 == 0:
    start = int(len(typeTxt)/2-25)
else:
    start = int(len(typeTxt)/2-25.5)
end = start+50
print(f'adj 길이의 중간 : {len(typeTxt)/2}, 시작 : {start}, 끝 : {end}')
```

```python
# 형용사 중립
li = []
for i in range(start,end):
    li.append(adj[i][1])
adj_neut = pd.DataFrame(li)
adj_neut.columns = ['ngram']
adj_neut['NEG'] = [0]*len(adj_neut)
adj_neut['NEUT'] = [1]*len(adj_neut)
adj_neut['POS'] = [0]*len(adj_neut)
adj_neut
```

```python
# 형용사 부정
li = []
for i in range(-1,-51,-1):
    li.append(adj[i][1])
adj_neg = pd.DataFrame(li)
adj_neg.columns = ['ngram']
adj_neg['NEG'] = [1]*len(adj_neg)
adj_neg['NEUT'] = [0]*len(adj_neg)
adj_neg['POS'] = [0]*len(adj_neg)
adj_neg
```

# 모든 단어 neg,neut,pos 분류한 데이터 프레임 생성

```python
# 모든 단어 긍정
li = []
for i in range(50):
    li.append(text[i][1])
text_pos = pd.DataFrame(li)
text_pos.columns = ['ngram']
text_pos['NEG'] = [0]*len(text_pos)
text_pos['NEUT'] = [0]*len(text_pos)
text_pos['POS'] = [1]*len(text_pos)
text_pos['ngram'] = text_pos['ngram'].apply(lambda x: x[:x.rfind('/')])
text_pos
```

```python
# 중립 데이터프레임 생성 위해 모든 단어의 중간 50개 데이터를 추출
typeTxt = text
if len(typeTxt)/2 % 2 == 0:
    start = int(len(typeTxt)/2-25)
else:
    start = int(len(typeTxt)/2-25.5)
end = start+50
print(f'text 길이의 중간 : {len(typeTxt)/2}, 시작 : {start}, 끝 : {end}')
```

```python
# 모든 단어 중립
li = []
for i in range(start,end):
    li.append(text[i][1])
text_neut = pd.DataFrame(li)
text_neut.columns = ['ngram']
text_neut['NEG'] = [0]*len(text_neut)
text_neut['NEUT'] = [1]*len(text_neut)
text_neut['POS'] = [0]*len(text_neut)
text_neut['ngram'] = text_neut['ngram'].apply(lambda x: x[:x.rfind('/')])
text_neut
```

```python
# 모든 단어 부정
li = []
for i in range(-1,-51,-1):
    li.append(text[i][1])
text_neg = pd.DataFrame(li)
text_neg.columns = ['ngram']
text_neg['NEG'] = [1]*len(text_neg)
text_neg['NEUT'] = [0]*len(text_neg)
text_neg['POS'] = [0]*len(text_neg)
text_neg['ngram'] = text_neg['ngram'].apply(lambda x: x[:x.rfind('/')])
text_neg
```

# knu 단어사전 불러오기

```python
knu = pd.read_csv('data/knu.csv',index_col='Unnamed: 0')
knu
```

# knu사전에 생성한 형용사, 명사, 동사, 모든 단어 데이터 프레임 이어붙이기

```python
dic = pd.concat([knu,adj_pos,adj_neut,adj_neg,
                 noun_pos,noun_neut,noun_neg,
                 verb_pos,verb_neut,verb_neg,
                 text_pos,text_neut,text_neg])
dic
```

```python
# 모델 데이터프레임 각 50개씩 잘 들어갔는지 확인
print(f'모델 데이터프레임 행 개수 : {(len(dic)-len(knu))/12}')
```

```python
# 중복 제거
dic = dic.drop_duplicates()
dic
```

```python
dic.to_csv('data/lexicon.csv',encoding='utf-8-sig',index=False)
```
